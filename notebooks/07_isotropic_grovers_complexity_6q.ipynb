# Increase in Grover's Algorithm complexity due to Isotropic Error

```python
from isotropic.algos.grover import get_grover_circuit
from isotropic.utils.state_transforms import generate_and_add_isotropic_error
from joblib import Parallel, delayed
from qiskit.quantum_info import Operator, Statevector

import jax.numpy as jnp
import matplotlib.pyplot as plt
```

## Oracle

We first implement the oracle that will add a phase to our desired search item ($000011$). 

**Note the negative sign on one of the diagonal entries.**

```python
oracle = jnp.eye(2**6).tolist()
oracle[3][3] = -1
U_w = Operator(oracle)
```

```python
num_qubits = 6
```

## Grover's Circuit

We can now build the circuit and visualise it

```python
grover_circuit = get_grover_circuit(num_qubits=num_qubits, U_w=U_w, iterations=4)
```

```python
grover_circuit.draw("mpl")
```

## Compare effect of different values of sigma

Instead of executing the circuit, we check to see the final statevector before measurements.

```python
statevector = Statevector(grover_circuit)
```

The probability of measuring the $000011$ state gives us a likelihood of success for our search exercise.

```python
print(
    f"Search success probability = {(statevector.probabilities_dict()['000011']) * 100: 4f} %"
)
```

## Perturbation with isotropic error

Let us now look at the effect of adding an isotropic error

```python
statevector_error = Statevector(
    generate_and_add_isotropic_error(statevector.data, sigma=0.9).tolist()
)
```

```python
print(
    f"Search success probability = {(statevector_error.probabilities_dict()['000011']) * 100: 4f} %"
)
```

We see that this has dropped to $81.6\%$ from $64.8\%$, a steeper drop than what we say for fewer qubits.

## Increased iterations

Now let's check the final probability (after isotropic error) when we run the algorithm with 5 iterations of the oracle + diffusion operator instead of 4.

```python
grover_circuit = get_grover_circuit(num_qubits=num_qubits, U_w=U_w, iterations=5)
grover_circuit.draw("mpl")
```

```python
statevector = Statevector(grover_circuit)
```

```python
statevector_error = Statevector(
    generate_and_add_isotropic_error(statevector.data, sigma=0.9).tolist()
)
```

```python
print(
    f"Search success probability = {(statevector_error.probabilities_dict()['000011']) * 100: 4f} %"
)
```

While we are not able to recover the original $81.6\%$ probability, we are able to bump it up to $76.5\%$ now at the cost of increased algorithmic complexity. We can try to see if it is possible to bump it higher by increasing the number of iterations to 6.

```python
grover_circuit = get_grover_circuit(num_qubits=num_qubits, U_w=U_w, iterations=6)
statevector = Statevector(grover_circuit)
statevector_error = Statevector(
    generate_and_add_isotropic_error(statevector.data, sigma=0.9).tolist()
)
print(
    f"Search success probability = {(statevector_error.probabilities_dict()['000011']) * 100: 4f} %"
)
```

We see that it is possible to trade off some more algorithmic complexity to get better success probability $(79\%$ instead of $76.5\%)$.

## Effect of error level

By tweaking the `sigma` parameter, we can increase the severity of the isotropic error. In this section, we take a look at how the success probability changes when we increase the error.

```python
def get_success_after_error(sigma):
    statevector_error = Statevector(
        generate_and_add_isotropic_error(statevector.data, sigma=sigma).tolist()
    )
    return statevector_error.probabilities_dict()["000011"] * 100
```

```python
sigmas = jnp.linspace(0.7, 0.99, 100)
```

```python
success = Parallel(n_jobs=8)(
    delayed(get_success_after_error)(sigma) for sigma in sigmas
)
```

```python
plt.plot(sigmas, success)
plt.xlabel("Sigma")
plt.ylabel("Success Probability (%)")
plt.show()
```
